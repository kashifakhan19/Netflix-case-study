```python

list_a = []

for i in range(len(df["show_id"])):
    show_id = df["show_id"][i]
    title = df["title"][i]
    type_i = df["type"][i]
    director = df["director"][i]
    cast = df["cast"][i]
    country = df["country"][i]
    date_added = df["date_added"][i]
    release_year = df["release_year"][i]
    rating = df["rating"][i]
    duration = df["duration"][i]
    genre = df["Genre"][i]
   
    if cast:
        cast_list = [x for x in cast.str.split(",")]
        for c in cast_list:
            list_a.append({
                "show_id": show_id,
                "title": title,
                "type": type_i,
                "director": director,
                "cast": c,
                "country": country,
                "date_added": date_added,
                "release_year": release_year,
                "rating": rating,
                "duration": duration,
                "Genre": genre    
            })
    else:
        list_a.append({
            "show_id": show_id,
            "title": title,
            "type": type_i,
            "director": director,
            "cast": None,
            "country": country,
            "date_added": date_added,
            "release_year": release_year,
            "rating": rating,
            "duration": duration,
            "Genre": genre    
        })


for i in list_a:
    print(i)
```

```python
df=pd.DataFrame(list_a)
df
```
